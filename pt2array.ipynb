```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
```

```python
# .ptファイルを読み込み
tensor = torch.load('density_3d_arrray/density_2410.pt')
middle_index=int((tensor.shape[1])/2)
tensor_mid=tensor[:,middle_index,:]
# テンソルをNumPy配列に変換
array_mid = tensor_mid.cpu().numpy()
np.save('density_2410_mid.npy', array_mid)
```

```python
def plot_showandsave(array):
    # ヒートマップを作成
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9, 6)) # 図の設定
        #c = ax.pcolor(array, cmap='binary_r') # ヒートマップ
        c = ax.pcolor(array, cmap='binary_r',norm=Normalize(vmin=0, vmax=3)) # ヒートマップ


        ax.set_aspect('equal', adjustable='box') # アスペクト比
        fig.colorbar(c, ax=ax) # カラーバー

plot_showandsave(np.rot90(array_mid,-1))
```
